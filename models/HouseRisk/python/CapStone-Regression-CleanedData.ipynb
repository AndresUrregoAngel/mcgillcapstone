```python
import pandas as pd

df_orig = pd.read_csv("./data/CleanedData8.csv") 
df_orig.head(n=5)
```

```python
df_orig['htype'] = df_orig.Type.astype('category').cat.codes
df_orig.head(n=5)
```

```python
df_orig['cityIndex'] = df_orig.City.astype('category').cat.codes
df_orig.head(n=5)
```

```python
df_orig['subtype'] = df_orig.Type.astype('category').cat.codes
df_orig.head(n=5)
```

```python
df_orig = df_orig.dropna()
df_orig["Rooms"] = df_orig['Rooms'].apply(lambda x: int(x.find(" Rooms")) if (x.strip() != "") else -1)
df_orig = df_orig[df_orig["Rooms"] != -1]
df_orig.head(n=5)
```

```python
df_input = df_orig.drop(['Type', 'type', 'City', 'lat', 'lng', 'Basement'], axis=1)
df_input.head(n=5)
```

```python
df_input1 = df_input.dropna()
df_input1.head(n=5)
```

```python
#cols_to_norm = ['Price']
#df_input1[cols_to_norm] = df_input1[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
#df_input1.head(n=5)
```

```python
X = df_input1.drop(['Risk Level'], axis=1)
y = df_input1['Risk Level']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor()  
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test) 
resultDF=pd.DataFrame({'Actual Calculated Risk Level':y_test, 'Predicted Risk Level':y_pred})  
resultDF.head(10)
X_test_pred = X_test.join(resultDF)
```

```python
from sklearn.metrics import confusion_matrix

X_test_high = X_test_pred[X_test_pred['Actual Calculated Risk Level'] >= 9.2]
X_test_high_pred_high = X_test_high[X_test_high['Predicted Risk Level'] >= 9.2]
print("actual high risk: ", X_test_high.shape[0], " predicate high risk: ", X_test_high_pred_high.shape[0])

X_pred_high = X_test_pred[X_test_pred['Predicted Risk Level'] >= 9.2]
print("total predicte high risk: ", X_pred_high.shape[0])

#type(y_test.transform(lambda v: v >= 9.2).values)
confusion_matrix(y_test.transform(lambda v: v >= 9.2).values, (lambda v: v >= 9.2)(y_pred))
```
